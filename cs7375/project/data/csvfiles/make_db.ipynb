```python
# Connect to your database
conn = sqlite3.connect("life_expectancy_model_data.db")
cursor = conn.cursor()

# Find all tables that start with "state_sex_le_p"
cursor.execute("""
    SELECT name 
    FROM sqlite_master 
    WHERE type='table' AND name LIKE 'state_sex_le_p%';
""")
tables = cursor.fetchall()

# Drop each table
for (table_name,) in tables:
    print(f"Dropping table: {table_name}")
    cursor.execute(f"DROP TABLE IF EXISTS {table_name}")

# Commit changes and close
conn.commit()
conn.close()
```
