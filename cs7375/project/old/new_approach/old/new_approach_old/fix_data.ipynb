```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('main_original.csv',encoding='latin-1')
df.head()
```

```python
#get rid of lines like Alabama	(blank)	NaN	75.5	75.2-77.5	0.0328
df_cleaned = df[~df.eq('(blank)').any(axis=1)]
df_cleaned.head(10)
```

```python
print(df_cleaned.shape)
print(df.shape)
```

```python
df_update = df_cleaned.drop(columns=['CensusTract', 'LifeExpectancyRange'])
print(df_update.shape)
```

```python
df_update.describe().T
```

```python
#based on above, will use mean
df_update['LifeExpectancy'] = df_update['LifeExpectancy'].fillna(df_update['LifeExpectancy'].mean())
df_update['LifeExpectancyStandardError'] = df_update['LifeExpectancyStandardError'].fillna(df_update['LifeExpectancyStandardError'].mean())
df_update.describe().T
```

```python
df_update.info()
```

```python
df_update['County'] = df_update['County'].str.split(',').str[0].str.strip()
df_update.info()
```

```python
print(df_update.head(10))
print(df_update.tail(10))
```

```python
df_update.to_csv('life_expectancy_by_county.csv', index=False)
```

```python
df2 = pd.read_csv('life_expectancy_by_county.csv',encoding='latin-1')
```

```python
df2.info()
```

```python
df2.describe()
```

```python
# Group by State and County, then average the numeric columns
df_avg = (
    df2.groupby(["State", "County"], as_index=False)
      .agg({
          "LifeExpectancy": "mean",
          "LifeExpectancyStandardError": "mean"
      })
)

# Save the cleaned dataset
#df_avg.to_csv("life_expectancy_by_county_cleaned.csv", index=False)

print("Original shape:", df.shape)
print("Cleaned shape:", df_avg.shape)
print(df_avg.head())
```

```python
df_avg.to_csv('life_expectancy_by_county_data.csv', index=False)
```
